```python
dv.plot_figure_three(nodes, G)
```

The three main descriptive characteristic of the network that the plots analyse are degree centrality, neighbour connectivity and the clustering coefficient.
* Degree centrality: based on the average connections of a person, males and females behave almost identically in this case, however the largest distinguishing factor is age. On average a 20-year-old has around 15 connections, however the older generation only have around 3-4 connections on average. Compared to the original article, we don't see a local maximum for the middle-aged generation in this case, as the users are less frequent on this network.
* Neighbour connectivity: our dataset suggests that the connections of male users have higher average degree throughout almost all ages, the biggest difference is between 20-35 years, which is the highest proportion of people included in our dataset. However, it can also be biased, as msot users were male, and a homophilic behaviour can be observed here.
* Clustering coefficient: the clustering coefficient is the highest for the youngest generation for men and for the eldest for female users. This shows contrary behaviour to the one in the article. Here, only female users increase their triadic closure when they get older, they usually tend to close their friend circles, whereas for younger boys they want to spread their friendships in the network.

When looking at the heatmap with the sum of rows as a normalizer, we can't observe that the proportion of same age connections is higher for same sex pairs, compared to mixed pairs and also, in general. On the other hand, the logarithm scaled values are reflecting a clearer picture for the graph.

The effect of intergenerational connections is greater for the female side, than for the male side, as can be seen for the higher values in the lower right corner of the heatmap. This can be reasoned by mothers being more socially active in the POKEC network, or other more deeply rooted causes. For men, the generational bond is stronger, so a certain age group will most probably will have connections within their own age group as well.

Male-female connections have their maxima at the younger age, however they have much higher values in the older generations, compared to same sex connections. This means, that most probably older married couples are making connections more frequently.
